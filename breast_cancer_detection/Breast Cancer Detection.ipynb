```python
from sklearn.model_selection import train_test_split

X = df.iloc[:, 1:].values
y = df.iloc[:, 0].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
def models(X_train, Y_train):
    
    #Logistic Regression
    from sklearn.linear_model import LogisticRegression
    log = LogisticRegression(random_state = 0)
    log.fit(X_train, y_train)
    
    #Decision Tree Classifier
    from sklearn.tree import DecisionTreeClassifier
    tree = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
    tree.fit(X_train, y_train)
    
    #Random Forest Classifier
    from sklearn.ensemble import RandomForestClassifier
    forest = RandomForestClassifier(n_estimators = 10,
                                    criterion = 'entropy',
                                    random_state = 0)
    forest.fit(X_train, y_train)
    
    print('Logistic Regression Training Accuracy: ', log.score(X_train, y_train))
    print('Decision Tree Classifier Training Accuracy: ', tree.score(X_train, y_train))
    print('Random Forest Classifier Training Accuracy: ', forest.score(X_train, y_train))
        
    return log,tree,forest
```

```python
model = models(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix

# y_pred0 = model[0].predict(X_test)
# y_pred1 = model[1].predict(X_test)
# y_pred2 = model[2].predict(X_test)
y_pred = []

for i in range(len(model)):
    y_pred.append(model[i].predict(X_test))

cm1 = confusion_matrix(y_test, y_pred[0])
cm2 = confusion_matrix(y_test, y_pred[1])
cm3 = confusion_matrix(y_test, y_pred[2])
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4)) 


sns.heatmap(cm1, annot=True, fmt='d', cmap='Blues', ax=axes[0])
axes[0].set_title("Confusion Matrix for Model 0")


sns.heatmap(cm2, annot=True, fmt='d', cmap='Greens', ax=axes[1])
axes[1].set_title("Confusion Matrix for Model 1")


sns.heatmap(cm3, annot=True, fmt='d', cmap='Reds', ax=axes[2])
axes[2].set_title("Confusion Matrix for Model 2")


plt.show()
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

for i in range(len(y_pred)):
    print(f"Accuracy for model {i}:",accuracy_score(y_test, y_pred[i]))
    print(classification_report(y_test, y_pred[i]))
```
